# Data Preprocessing Template

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Position_Salaries_1100.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

# training the linear model

```
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x , y)
```

# visualizing the linear mode

```
plt.scatter(x , y , color = 'red')
plt.plot(x , lin_reg.predict(x) , color = 'blue')
plt.xlabel('level')
plt.ylabel('salary')
```

# training the polynomial regression

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 5)
x_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly , y)
```

# visualizing the polynomial model

```
plt.scatter(x , y , color = 'red')
plt.plot(x , lin_reg_2.predict(x_poly) , color = 'blue')
plt.xlabel('level')
plt.ylabel('salary')
```

# Predicting a new result with Linear Regression

```
lin_reg.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```
lin_reg_2.predict(poly_reg.fit_transform([[4.2]]))
```

# calculating the r2_score

```
from sklearn.metrics import r2_score

y_pred = lin_reg_2.predict(x_poly)
r2 = r2_score(y, y_pred)
print("R2 Score:", r2)
```
